```python
# Needed imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Preparing for all steps of the algorithm
np.random.seed = 42
n_samples = 10000
mean = [0, 0]
cov = [[1, 0], [0, 1]]
data = np.random.multivariate_normal(mean, cov, n_samples)
kmeans = KMeans(n_clusters=3)
averages = []
iters = 10000
target_accuracy = 1e-9
stabilized_iteration = None
```

```python
# Algorithm itself
for i in range(iters):
  kmeans.fit(data)
  centroids = kmeans.cluster_centers_
  norms = np.linalg.norm(centroids, axis=1)
  avg = np.mean(norms)
  averages.append(avg)
cumulative = np.cumsum(averages) / np.arange(1, iters + 1)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(cumulative)
plt.title('Cumulative Average of Centroid Norms')
plt.xlabel('Iteration')
plt.ylabel('Cumulative Average')
plt.grid()
plt.axhline(y=cumulative[-1], color='r', linestyle='--', label='Optimal (final) average')
plt.legend()
plt.show()
```

```python
# Last task
for i in range(1, len(cumulative)):
    difference = abs(cumulative[i] - cumulative[i - 1])
    print(difference)
    if difference < target_accuracy:
        stabilized_iteration = i + 1
        break
if stabilized_iteration:
    print(f"The cumulative average stabilized to 9-digit accuracy at iteration {stabilized_iteration}.")
else:
    print("The cumulative average did not reach 9-digit accuracy within the specified iterations.")
```
